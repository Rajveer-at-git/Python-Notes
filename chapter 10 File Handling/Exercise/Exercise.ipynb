```python
from pathlib import Path
path = Path('guest_books.txt')

guest_book = []
name = ''

while 1:
    name = input("Enter your name: ")
    if name == 'exit':
        break
    guest_book.append(name)

name = ''
for guest in guest_book:
    name += f"{guest}\n"

path.write_text(name)
```

```python
from pathlib import Path 
import json

def write_fav_no(path,fav_no):
    fav_no_json = json.dumps(fav_no)
    path.write_text(fav_no_json)

def get_fav_no():
    fav_no = input("Enter your favorite no.: ")
    return fav_no

def read_fav_no(path):
    content = path.read_text()
    fav_num = json.loads(content)
    return fav_num

path = Path('favorite_num.json')
fav_no = get_fav_no()
write_fav_no(path,fav_no)
read_fav_num = read_fav_no(path)
print(f"Okay so your favroite number is {read_fav_num}.")
```

```python
from pathlib import Path
import json


def get_stored_user_info(path):
    if path.exists():
        content = path.read_text()
        user_info = json.loads(content)
        return user_info
    else:
        return None

def get_new_user_info(path):
    user_info = {"name": None, "age": None, "gender": None}
    user_info["name"] = input("Enter your name: ")
    user_info["age"] = input("Enter your age: ")
    user_info["gender"] = input("Enter your gender: ")
    content = json.dumps(user_info)
    path.write_text(content)
    return user_info

def greet_user():
    path = Path('user_info.json')
    user_info = get_stored_user_info(path)

    is_correct = False
    if user_info:
        print(f"Is the User name: '{user_info['name']}' correct?")
        user_input = input("Yes(y) or No(n)? ").lower()
        if user_input in ['y', 'yes']:
            is_correct = True

    if is_correct: 
        print(f"Welcome back {user_info['name']}!")
    else:
        user_info = get_new_user_info(path)
        print(f"So your details are:\nName: {user_info['name']}\nAge: {user_info['age']}\nGender: {user_info['gender']}")
    
    
greet_user()


```
