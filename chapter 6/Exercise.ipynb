```python
rivers = {'Ganga':'India', 'Nile' : 'Egypt', 'Amazon' : 'Brazil'}
for river,country in rivers.items():    # Rivers and their Country
    print(f"{river} runs through {country}.")

print(f'\nThe Rivers taken are:')
for river in rivers:    # Name of rivers
    print(f'\t{river}')

print(f'Countries taken are:')
for country in rivers.values():
    print(f'\t{country}')
```

```python
cities = {
    'indore' : {
            'country': 'india',
            'population': 100000,
            'fact': 'cleanest city of India'
    },

    'new york' : {
            'country': 'USA',
            'population': 1000000,
            'fact': 'Richest city in the world'
    },

    'tokyo' : {
            'country': 'japan',
            'population': 10000000,
            'fact': 'biggest city by population'
    }
}

for city, city_info in cities.items():
    print(f"City: {city}")
    print(f"\tCountry: {city_info['country']}")
    print(f"\tPopulation: {city_info['population']}")    
    print(f"\tFact: {city_info['fact']}")
```
