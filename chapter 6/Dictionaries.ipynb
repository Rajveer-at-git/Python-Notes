```python
# get() method does not throws an error it returns 'None'(or the value we specified) if the key doesn't exists
# in the dictionary
data = {'name': 'Guts', 'anime': 'Berserk'}

# Scenario 1: The key exists
print(data.get('anime'))
# Output: Berserk

# Scenario 2: The key does NOT exist (Default behavior)
print(data.get('age'))
# Output: None (No error, the code keeps running)

# Scenario 3: The key does NOT exist (Custom fallback)
print(data.get('age', 'Unknown'))
# Output: Unknown
```

```python
# Nesting
# List of dictionaries
alien_0 = {'color' : 'green', 'points' : 5}
alien_1 = {'color' : 'yellow', 'points' : 10}
alien_2 = {'color' : 'red', 'points' : 15}

aliens = [alien_0, alien_1, alien_2]
for alien in aliens:
    print(alien)
```

```python
# Practical example
# create 30 aliens and store them in aliens list
aliens = []

for alien in range(30):
    new_alien = {'color': 'green', 'points': 5, 'speed': 'slow'}
    aliens.append(new_alien)

# Show first five aliens in the list
for alien in aliens[:5]:
    print(alien)
print('...')

# How many aliens are in the list
print(f"No. of aliens: {len(aliens)}")
```

```python
# Now changing the first 3 aliens to yellow
aliens = []

for alien in range(30):
    new_alien = {'color': 'green', 'points': 5, 'speed': 'slow'}
    aliens.append(new_alien)

for alien in aliens[:3]:
    if(alien['color']  == 'green'):
        alien['color'] = 'yellow'
        alien['points'] = 10
        alien['speed'] = 'medium'

# Show first five aliens in the list
for alien in aliens[:5]:
    print(alien)
print('...')

# How many aliens are in the list
print(f"No. of aliens: {len(aliens)}")
```

```python
# And For changing the yellow aliens to red
aliens = []

for alien in range(30):
    new_alien = {'color': 'yellow', 'points': 10, 'speed': 'medium'}
    aliens.append(new_alien)

for alien in aliens[:3]:
    if(alien['color']  == 'green'):
        alien['color'] = 'yellow'
        alien['points'] = 10
        alien['speed'] = 'medium'

    elif(alien['color'] == 'yellow'):
        alien['color'] = 'red'
        alien['points'] = 15
        alien['speed'] = 'fast'

# Show first five aliens in the list
for alien in aliens[:5]:
    print(alien)
print('...')

# How many aliens are in the list
print(f"No. of aliens: {len(aliens)}")
```

```python
# Example-2
# Dictionary of lists
favorite_languages = {
'jen': ['python', 'rust'],
'sarah': ['c'],
'edward': ['rust', 'go'],
'phil': ['python', 'haskell'],
}
for name,languages in favorite_languages.items():
    print(f"{name.title()}'s favorite languages are:")

    for language in languages:
        print(f"{language.title()}")
    print('\n')
```

```python
# Dictionaries inside dictionary
users = {
    'aeinstein': {
        'first': 'albert',
        'last': 'einstein',
        'location': 'princeton',
        },
    'mcurie': {
        'first': 'marie',
        'last': 'curie',
        'location': 'paris',
        },
}
for username, user_info in users.items():

    print(f"\nUsername: {username}")

    full_name = f"{user_info['first']} {user_info['last']}"
    location = user_info['location']

    print(f"\tFull name: {full_name.title()}")
    print(f"\tLocation: {location.title()}")
```
