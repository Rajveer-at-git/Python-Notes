Reading From a file

```python
from pathlib import Path

path = Path('pi_digits.txt')
contents = path.read_text()
contents = contents.rstrip()
print(contents)
```

```python
from pathlib import Path

path = Path('pi_digits.txt')
contents = path.read_text().rstrip()

lines = contents.splitlines()
for line in lines:
    print(line)
```

```python
from pathlib import Path

path = Path('pi_digits.txt')
content = path.read_text().rstrip()
lines = content.splitlines()
pi_string = ''

for line in lines:
    pi_string += line
    print(pi_string)
    print(len(pi_string))
```

```python
# we can use lstrip to remove the spaces in between
from pathlib import Path

path = Path('pi_digits.txt')
content = path.read_text().rstrip()
lines = content.splitlines()
pi_string = ''

for line in lines:
    pi_string += line.lstrip()
    print(pi_string)
    print(len(pi_string))
```

```python
# we can also print only first 5 or 10 decimal values
from pathlib import Path

path = Path('pi_digits.txt')
content = path.read_text().rstrip()
lines = content.splitlines()
pi_string = ''

for line in lines:
    pi_string += line.lstrip()

print(pi_string[:7])
print(pi_string[0:12])
```

```python
from pathlib import Path
path = Path('pi_million_digits.txt')

content = path.read_text().rstrip()
lines = content.splitlines()
for line in lines:
    print(len(line))

print(len(content))
print(len(lines))
```

```python
from pathlib import Path

path = Path('pi_million_digits.txt')
contents = path.read_text()
lines = contents.splitlines()
pi_string = ''

for line in lines:
    pi_string += line.strip()

birthday = input("Enter your birthday, in the form mmddyy: ")
if birthday in pi_string:
    print("Your birthday appears in the first million digits of pi!")
else:
    print("Your birthday does not appear in the first million digits of pi.")
```

```python
# We can replace any word from a string using .replace() method
str = 'I Like cats.'
new_str = str.replace('cats','dogs')
print(new_str)
```

Writing To a File

```python
from pathlib import Path

path = Path('programming.txt')
path.write_text('I Love Programming')
```

```python
print(path.read_text())
```

NOTE: .write_text() firstly erases all the content from the file then writes to it

```python
# Writing multiple programs

from pathlib import Path

contents = "I love programming.\n"
contents += "I love creating new games.\n"
contents += "I also love working with data.\n"

path = Path('programming.txt')
path.write_text(contents)
```
