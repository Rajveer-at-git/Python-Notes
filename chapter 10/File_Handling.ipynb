```python
from pathlib import Path

path = Path('pi_digits.txt')
contents = path.read_text().rstrip()

lines = contents.splitlines()
for line in lines:
    print(line)
```

```python
from pathlib import Path

path = Path('pi_digits.txt')
content = path.read_text().rstrip()
lines = content.splitlines()
pi_string = ''

for line in lines:
    pi_string += line
    print(pi_string)
    print(len(pi_string))
```

```python
# we can use lstrip to remove the spaces in between
from pathlib import Path

path = Path('pi_digits.txt')
content = path.read_text().rstrip()
lines = content.splitlines()
pi_string = ''

for line in lines:
    pi_string += line.lstrip()
    print(pi_string)
    print(len(pi_string))
```

```python
# we can also print only first 5 or 10 decimal values
from pathlib import Path

path = Path('pi_digits.txt')
content = path.read_text().rstrip()
lines = content.splitlines()
pi_string = ''

for line in lines:
    pi_string += line.lstrip()

print(pi_string[:7])
print(pi_string[0:12])
```

```python
from pathlib import Path

path = Path('pi_million_digits.txt')
contents = path.read_text()
lines = contents.splitlines()
pi_string = ''

for line in lines:
    pi_string += line.strip()

birthday = input("Enter your birthday, in the form mmddyy: ")
if birthday in pi_string:
    print("Your birthday appears in the first million digits of pi!")
else:
    print("Your birthday does not appear in the first million digits of pi.")
```

Exception Handling

```python
try:
    print(5/0)
except ZeroDivisionError:
    print("You can't divide by zero!")
```

```python
print("Enter two nos. and I'll Divide them.")
print("Enter q to quit")

while 1:
    first_number = input("First number: ")
    if(first_number == 'q'):
        break

    second_number = input("Second number: ")
    if(second_number == 'q'):
        break
    
    try:
        answer = int(first_number)/int(second_number)
    except ZeroDivisionError:
        print("Cannot divide by 0.")
    else:
        print(f'Answer: {answer}')
```

```python
# Sometimes we want to silently pass the error without showing it to the user
# for that use pass
from pathlib import Path


def count_words(path):
    try:
        contents = path.read_text(encoding='utf-8') # It can give error
    except FileNotFoundError:
        pass
    else:    
        words = contents.split()    # breaks after a space
        num_words = len(words)
        print(f"The file {path} has approximate {num_words} words.") # Total no. of words in the file

filenames = ['alice.txt','siddhartha.txt','moby_dick.txt','little_women.txt']   # Note: siddhartha.txt doesn't exist
for file in filenames:
    path = Path(file)
    count_words(path)
```
