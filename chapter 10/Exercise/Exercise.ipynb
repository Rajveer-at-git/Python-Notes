Exercise 10.2

```python
from pathlib import Path
path = Path('guest_books.txt')

guest_book = []
name = ''

while 1:
    name = input("Enter your name: ")
    if name == 'exit':
        break
    guest_book.append(name)

name = ''
for guest in guest_book:
    name += f"{guest}\n"

path.write_text(name)
```
