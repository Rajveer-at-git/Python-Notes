Exercise 10.2

```python
from pathlib import Path
path = Path('guest_books.txt')

guest_book = []
name = ''

while 1:
    name = input("Enter your name: ")
    if name == 'exit':
        break
    guest_book.append(name)

name = ''
for guest in guest_book:
    name += f"{guest}\n"

path.write_text(name)
```

```python
print("Enter two no.s and I'll give sum of them")

num1 = input('First no.: ')
num2 = input('Second no.: ')

try:
    result = int(num1) + int(num2)
except ValueError:
    print("Please enter numeric values only.")
else:
    print(f'Sum: {result}')
```

Exercise 10.5

```python
from pathlib import Path

animals = ['cats.txt','dogs.txt']

try:
    for animal in animals:
        path = Path(animal)
        print(path.read_text().rstrip())
except FileNotFoundError:
    print(f"File {path} Not found!")
    
```
