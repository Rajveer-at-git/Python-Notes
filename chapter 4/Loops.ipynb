```python
for _ in range(10):
    print('Hello!') # The _ is ignored, by standard convention
```

```python
# for Loop
magicians = ["Mage_A","Mage_B","Mage_C"]
for mage in magicians:
    print(mage)
```

```python
squares = []
for value in range(1,11):
    squares.append(value ** 2)
print(squares)
```

```python
squares = []        # Same functioning using insert
for value in range(1,11):
    squares.insert(10,value ** 2)
print(squares)
```

```python
digits = list(range(1,11))
print(min(digits))
print(max(digits))
print(sum(digits))
```

```python
# List comprehension
squares = [value**2 for value in range(1,11)]
print(squares)
```

```python
# Slicing the List
anime = ['Naruto','Gintama','Dragon ball','Beyblade']
print(anime[0:3])
print(anime[:3])
print(anime[2:])
print(anime[:-3])
print(anime[-3:])
print(anime[1:-1])
print(anime[0:3:2])
```

```python
players = ['charles', 'martina', 'michael', 'florence', 'eli']
print("Here are the first three players on my team:")
for player in players[:3]:
    print(player.title())
```

```python
# Copying Lists
my_foods = ['pizza', 'falafel', 'carrot cake']
friend_foods = my_foods[:]
# NOTE: if we write 
# friend_foods = my_foods , any changes made to my_foods will be reflected to friend_foods too cause 
# both are working on a single list
# To over come this we add [:] for copying the list
print("My favorite foods are:")
print(my_foods)
print("\nMy friend's favorite foods are:")
print(friend_foods)
```

```python
my_foods = ['pizza', 'falafel', 'carrot cake']
friend_foods = my_foods[:]

my_foods.append('cannoli')
friend_foods.append('ice cream')

print(my_foods)
print(friend_foods)
```

```python
dimensions = (200, 50)  # Although we can't change the value in a tuple but we can assign a whole new value
# to the tuple like this
print("Original dimensions:")
for dimension in dimensions:
    print(dimension)
dimensions = (400, 100)
print("\nModified dimensions:")
for dimension in dimensions:
    print(dimension)
```
