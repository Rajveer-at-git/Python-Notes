```python
users = ['Matthew','John','Admin','hinata']
if users:
    for user in users:
        if(user == 'Admin'):
            print(f"Hello {user}. Would you like to see today's status report")
        else:
            print(f"Greetings {user}. Thank you for logging in.")
else:
    print(f"No logins.")
```

```python
users = ['Matthew','John','Admin','Hinata','Robin']
current_users = [user.lower() for user in users]    # List comprehension
new_users = ['Naruto','Sasuke','Luffy','Hinata','Robin']
for new_user in new_users:
    if new_user.lower() in current_users:
        print(f"This username had already been taken, please choose different name.")
    else:
        print(f"This username is avaliable.")
    
```

```python
nums = list(range(1,10))
for num in nums:
    if(num == 1):
        print(f"Num: {num}st")
    elif(num == 2):
        print(f"Num: {num}nd")
    elif(num == 3):
        print(f"Num: {num}rd")
    else:
        print(f"Num: {num}th")
```
