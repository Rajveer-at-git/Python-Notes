```python
prompt = "\nEnter any thing and I'll repeat it for you."
prompt += '\nEnter "quit" for stop.'

message = prompt
while(message != 'quit'):
    message = input(prompt)
    if message != 'quit':
        print(message)
```

```python
# using a variable called flag to check the loop condition
prompt = "\nEnter any thing and I'll repeat it for you."
prompt += '\nEnter "quit" for stop.'

active = True
while active:
    message = input(prompt)
    
    if message == 'quit':
        active = False
    else:
        print(message)
```

```python
# using continue 
num = 0
while(num<10):
    num += 1
    if(num % 2 == 0):
        continue
    print(num)
    
```

```python
unconfirmed_users = ['goku','naruto','vegeta','tanjiro']
confirmed_users = []

while unconfirmed_users:
    users = unconfirmed_users.pop()

    print(f"Confirming user {users.title()}")
    confirmed_users.append(users)

print(f"\nFollowing users have been confirmed:")
for user in confirmed_users:
    print(user.title())
```

```python
pets = ['dog', 'cat', 'dog', 'goldfish', 'cat', 'rabbit', 'cat']
print(pets)

while 'cat' in pets:
    pets.remove('cat')

print(pets)
```

```python
responses = {}
loop_active = True

while loop_active:
    name = input("Enter your name: ")
    response = input("Name the country you wanna visit: ")

    responses[name] = response

    person = input("Anyone else wanna take poll?(y/n)")
    if(person == 'n'):
        loop_active = False

for name,response in responses.items():
    print(f"{name} : {response}")
```
