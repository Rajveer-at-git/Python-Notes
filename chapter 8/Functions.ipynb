```python
def get_name(first_name, last_name):
    person = f'{first_name} {last_name}'
    return person.title()

name = get_name('rajveer','choudhary')
print(name)
```

```python
# Making an argument optional, Note the optional parameter comes at the last
def get_formatted_name(first_name, last_name, middle_name=''):
    """Return a full name, neatly formatted."""

    if middle_name:
        full_name = f"{first_name} {middle_name} {last_name}"

    else:
        full_name = f"{first_name} {last_name}"

    return full_name.title()

musician = get_formatted_name('jimi', 'hendrix')
print(musician)
musician = get_formatted_name('john', 'hooker', 'lee')
print(musician)
```

```python
def build_person(first_name, last_name):
    """Return a dictionary of information about a person."""

    person = {'first': first_name, 'last': last_name}
    return person

musician = build_person('jimi', 'hendrix')
print(musician)
```

```python
def person(first_name, last_name, age=None):
    guy = {'first' : first_name, 'last' : last_name}
    if age:
        guy['age'] = age
    return guy

person1 = person('Rajveer','Chourasiya',19)
print(person1)

```

```python
def get_formatted_name(first_name, last_name):
    """Return a full name, neatly formatted."""
    full_name = f"{first_name} {last_name}"
    return full_name.title()

while True:
    print("\nPlease tell me your name:")
    print("(enter 'q' at any time to quit)")
    f_name = input("First name: ")
    if f_name == 'q':
        break
    l_name = input("Last name: ")
    if l_name == 'q':
        break
    formatted_name = get_formatted_name(f_name, l_name)
    print(f"\nHello, {formatted_name}!")
```

```python
# Creating Modules and importing fns from those modules
# Create one file named pizza.py 
# Then write one fn name "make_pizza" in it, call that fn here and pass the required parameters

import pizza
pizza.make_pizza(14,'extra cheese','pepperoni')
```

```python
# We can also import specific fns from a module instead of importing whole module
# Syntax: from module_name import function_name
# or      from module_name import function_0, function_1, function_2

from pizza import make_pizza
make_pizza(16, 'pepperoni')
make_pizza(12, 'mushrooms', 'green peppers', 'extra cheese')
```

```python
# we can also give the fn or whole module an alias name using 'as'
import pizza as p   # module alias
p.make_pizza(16, 'pepperoni')
p.make_pizza(12, 'mushrooms', 'green peppers', 'extra cheese')
```

```python
from pizza import make_pizza as mp  # fn alias
mp(16, 'pepperoni')
mp(12, 'mushrooms', 'green peppers', 'extra cheese')
```

```python
# To import all fns from a module
from pizza import *
make_pizza(16, 'pepperoni')
make_pizza(12, 'mushrooms', 'green peppers', 'extra cheese')
```
