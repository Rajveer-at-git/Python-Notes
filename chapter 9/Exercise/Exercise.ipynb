```python
from random import randint
class Die:
    def __init__(self,sides=6):
        self.sides = sides

    def roll_die(self):
        for _ in range(10):
            print(randint(1, self.sides))


dice_6 = Die()
dice_10 = Die(10)
dice_20 = Die(20)

print("Six-sided dice rolled 10 times:")
dice_6.roll_die()
    
print("\nTen-sided dice rolled 10 times: ")
dice_10.roll_die()

print("\nTwenty-sided dice rolled 20 times: ")
dice_20.roll_die()
```
