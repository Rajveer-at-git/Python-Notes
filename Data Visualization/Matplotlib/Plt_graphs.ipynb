```python
import matplotlib.pyplot as plt

input_values = [1, 2, 3, 4, 5]
squares = [1,4,9,16,25]

plt.style.use('seaborn-v0_8')
fig, ax = plt.subplots()
ax.plot(input_values, squares, linewidth = 3)

# Set chart title and label axes
ax.set_title("Square Numbers", fontsize=20)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

# Set size of tick labels
ax.tick_params(labelsize=10)

plt.show()
```

```python
# For working with series of data
import matplotlib.pyplot as plt

# We pass x and y values in the form of list
x_values = [1, 2, 3, 4, 5]
y_values = [1, 4, 9, 16, 25]

plt.style.use('seaborn-v0_8')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, s=100)

# Set chart title and label axes.
ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

# Set size of tick labels.
ax.tick_params(labelsize=14)
plt.show()
```

```python
import matplotlib.pyplot as plt

x_values = range(1, 1001)
y_values = [x**2 for x in x_values] # List comprehension
plt.style.use('seaborn-v0_8')
fig, ax = plt.subplots()
plt.scatter(x_values, y_values, s=10, color='red')  # state-based (implicit) interface

# Set chart title and label axes.
ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

# Set the range for each axis.
# Format: ax.axix([x_min, x_max, y_min, y_max])
ax.axis([0, 1100, 0, 1_100_000])    # Underscore is just syntax
ax.ticklabel_format(style='plain')

plt.show()
```

- We can also use rgb colors for creating custom colors, like this:

    ax.scatter(x_values, y_values, color=(0, 0.8, 0), s=10)

- values closer to 0 creates darker colors, while colors closer to 1 are bright

##### Color Maps


```python
import matplotlib.pyplot as plt

x_values = range(1, 1001)
y_values = [x**2 for x in x_values] # List comprehension
plt.style.use('seaborn-v0_8')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, c=y_values, cmap=plt.cm.Reds, s=10) # object-oriented based (explicit) interface

# Set chart title and label axes.
ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

# Set the range for each axis.
# Format: ax.axix([x_min, x_max, y_min, y_max])
ax.axis([0, 1100, 0, 1_100_000])    # Underscore is just syntax
ax.ticklabel_format(style='plain')

plt.show()
```
