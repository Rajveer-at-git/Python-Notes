Github's API call for asking info about the most popular python repos.
https://api.github.com/search/repositories?q=language:python+sort:stars

https://api.github.com/ -> part of Github that responds to API calls

search/repositories -> to conduct a search through all repos in github

? ->about to start an argument

q= -> q means query, q= lets us begin specifying a query

language:python -> primary language should be python

+sort:stars -> sort ascending to the star marked on the repo

```python
import requests

# Make an API call and check the response
url = "https://api.github.com/search/repositories"
url += "?q=language:python+sort:stars:>10000"

headers = {"Accept": "application/vnd.github.v3+json"}  # Use third version of API
r = requests.get(url, headers=headers)
print(f"Status code: {r.status_code}")

# Convert the response to a dicitionary
response_dict = r.json()

# Process results
print(response_dict.keys())
```
