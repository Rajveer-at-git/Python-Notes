```python
"""Mark the first and the last point in the graph"""


import matplotlib.pyplot as plt

from random_walk import RandomWalk

# Keep making new walks as long as program is running
while True:
    rw = RandomWalk(5000)
    rw.fill_walk()

    # Plot the points in the walk.
    plt.style.use('classic')
    fig, ax = plt.subplots(figsize=(8,8), dpi=128)
    point_numbers = range(rw.num_points)
    ax.plot(rw.x_values, rw.y_values, linewidth = 3)
    ax.set_aspect('equal')

    # Remove the axes.
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    plt.show()

    keep_running = input("Make another walk? (y/n): ")
    if keep_running == 'n':
        break

```

```python
"""Plot a graph on the rainfall received in Sitka(temperate rainforest)"""

import csv
import matplotlib.pyplot as plt
from pathlib import Path
from datetime import datetime

path = Path("sitka_weather_2021_full.csv")
lines = path.read_text().splitlines()

reader = csv.reader(lines)
header_row = next(reader)

prcps, dates = [], []
for row in reader:
    current_date = datetime.strptime(row[2], '%Y-%m-%d')
    prcp = float(row[5])
    dates.append(current_date)
    prcps.append(prcp)

plt.style.use('seaborn-v0_8')
fig, ax = plt.subplots()
ax.plot(dates, prcps)
ax.set_title("Rainfall received in Sitka Highs", fontsize=20)
ax.set_xlabel("", fontsize=16)
fig.autofmt_xdate()
ax.set_ylabel("PRCP", fontsize=16)
plt.show()

```

```python
import matplotlib.pyplot as plt
from datetime import datetime
import csv
from pathlib import Path

path = Path("Detroid Temperature dataset.csv")
lines = path.read_text().splitlines()

reader = csv.reader(lines)
header_row = next(reader)

dates, highs, lows = [], [], []
for row in reader:
    current_date = datetime.strptime(row[2],'%Y-%m-%d')
    try:
        high = int(row[4])
        low = int(row[5])
    except ValueError:
        print(f"Missing data at date {current_date}")
    else:
        dates.append(current_date)
        highs.append(high)
        lows.append(low)

# Plot the data
plt.style.use('seaborn-v0_8')
fig, ax = plt.subplots()
ax.plot(dates, highs, color='red', alpha=0.5)
ax.plot(dates, lows, color='blue', alpha=0.5)
ax.fill_between(dates, highs, lows, color='green', alpha=0.5)
fig.autofmt_xdate()

# Set the labels
ax.set_title("High and Low Temperature in Detroid", fontsize=20)
ax.set_xlabel("", fontsize=16)
ax.set_ylabel("Temperature", fontsize=16)
plt.show()
```

```python
from operator import itemgetter
import plotly.express as px
import requests

# Make an API call and check the response.
url = "https://hacker-news.firebaseio.com/v0/topstories.json"
r = requests.get(url)
print(f"Status code: {r.status_code}")

# Process information about each submission.
articles = r.json()
submission_dicts = []

for article in articles[:10]:
    url = f"https://hacker-news.firebaseio.com/v0/item/{article}.json"
    # https://hacker-news.firebaseio.com/v0/item/48526661.json 
    r = requests.get(url)
    art_details = r.json()
    
    # We used .get() cause its allows for passing a second default argument
    # which will be place when no data is avaliable
    # This saves us of writing a four line try-except block
    # It is mostly used with dictionaries
    submission_dicts.append({
        'title': art_details.get('title', 'Unknown'),
        'comments': art_details.get('descendants', 0)
    })
    

submission_dicts = sorted(submission_dicts, key=itemgetter('comments'), reverse=True)

article_title = [item['title'] for item in submission_dicts]
comments = [item['comments'] for item in submission_dicts]

# Make visualization.
title = "Most Popular Articles on Hacker News"
labels = {'x': 'Article', 'y': 'Comments'}
hover_texts = [item['title'] for item in submission_dicts]

# Create a new list of short titles for the x-axis and cut off after 15 characters
short_titles = [title[:15] + "..." if len(title) > 15 else title for title in hover_texts]

# Update your figure to use the short titles on X, and full titles on hover
fig = px.bar(x=short_titles, y=comments, title=title, labels=labels, hover_name=hover_texts)
fig.update_layout(title_font_size=28, xaxis_title_font_size=20,
                yaxis_title_font_size=20)
fig.show()
```
